```python
def appendSpherical_np(xyz):
    ptsnew = np.hstack((xyz, np.zeros(xyz.shape)))
    xy = xyz[:,0]**2 + xyz[:,1]**2
    ptsnew[:,3] = np.sqrt(xy + xyz[:,2]**2) #range
    #ptsnew[:,4] = np.arctan2(np.sqrt(xy), xyz[:,2]) # for elevation angle defined from Z-axis down
    ptsnew[:,4] = np.arctan2(xyz[:,2], np.sqrt(xy)) # for elevation angle defined from XY-plane up
    ptsnew[:,5] = np.arctan2(xyz[:,1], xyz[:,0]) #azimuth
    return ptsnew

def sph2cart(az, el, r):
    rcos_theta = r * np.cos(el)
    x = rcos_theta * np.cos(az)
    y = rcos_theta * np.sin(az)
    z = r * np.sin(el)
    return x, y, z

def convex_hull_volume(pts):
    ch = spatial.ConvexHull(pts)
    dt = spatial.Delaunay(pts[ch.vertices])
    tets = dt.points[dt.simplices]
    return np.sum(tetrahedron_volume(tets[:, 0], tets[:, 1],
                                     tets[:, 2], tets[:, 3]))

def tetrahedron_volume(a, b, c, d):
    return np.abs(np.einsum('ij,ij->i', a-d, np.cross(b-d, c-d))) / 6

def moving_average(a, n=3):
    """
    return the moving average of a 1D vector.
    total length is preserved by using same values on edges.

    Parameter:
    ----------
    a: ndarray (1,n)
        vector to apply moving average to
    n: int
        window size

    Returns:
    ----------
    ndarray (1,n)
        a with moving average filter applied
    """
    return np.convolve(a, np.ones((n,)) / n, mode="same")
```

```python
stl_ffn = '/home/meso/data/yalboroo_collection/hailstone09.stl'
#stl_ffn = '/home/meso/data/brisbane_collection/hailstone_11.stl'
# load mesh
mymesh = trimesh.load_mesh(stl_ffn)
# volumetric center of mass which we can set as the origin for our mesh
mymesh.vertices -= mymesh.center_mass
mymesh_convex_hull = mymesh.convex_hull

# Naive way of finding the best pair in O(H^2) time if H is number of points on hull
hdist = cdist(mymesh_convex_hull.vertices, mymesh_convex_hull.vertices, metric='euclidean')
Dmax = hdist.max()
# Get the farthest apart points
bestpair = np.unravel_index(hdist.argmax(), hdist.shape)
Dmax_points = [mymesh_convex_hull.vertices[bestpair[0]], mymesh_convex_hull.vertices[bestpair[1]]]
#calculate vector between Dmax points
Dint_Dmin_plane_nvec = Dmax_points[0] - Dmax_points[1]
Dmax_midpoint = (Dmax_points[0] + Dmax_points[1])/2

#slice convex hull along Dint/Dmin plane
Dint_Dmin_myslice = mymesh.section(plane_origin=Dmax_midpoint, 
                        plane_normal=Dint_Dmin_plane_nvec)

slice_2D, to_3D = Dint_Dmin_myslice.to_planar()
#order 2D slice points by their azimuth
azimuth = np.arctan2(slice_2D.vertices[:,0], slice_2D.vertices[:,1])
sort_ind = np.argsort(azimuth)
slice_2D.vertices = slice_2D.vertices[sort_ind, :]
#calculate range (to find minimums
x_mean = np.mean(slice_2D.vertices[:,0])
y_mean = np.mean(slice_2D.vertices[:,1])
Dint_Dmin_range = np.sqrt((slice_2D.vertices[:,0] - x_mean)**2 + (slice_2D.vertices[:,1] - y_mean)**2)
#smooth
Dint_Dmin_range_smooth = moving_average(Dint_Dmin_range, 15)
#find peaks
Dint_Dmin_peaks_ind, prop = find_peaks(-Dint_Dmin_range_smooth)
#MBB
Dint_Dmin_bounding_box_stats = mbb.MinimumBoundingBox(slice_2D.vertices[Dint_Dmin_peaks_ind,:])

fig = plt.figure(figsize=[8,8])
plt.scatter(slice_2D.vertices[:,0], slice_2D.vertices[:,1], c=Dint_Dmin_range)
for ind in Dint_Dmin_peaks_ind:
    plt.plot(slice_2D.vertices[:,0][ind], slice_2D.vertices[:,1][ind], 'r.')
mbb_points = np.array(list(Dint_Dmin_bounding_box_stats.corner_points))
plt.plot(mbb_points[:,0], mbb_points[:,1], 'b.')







Dint_hdist = cdist(Dint_Dmin_myslice.vertices, Dint_Dmin_myslice.vertices, metric='euclidean')
Dint_bestpair = np.unravel_index(Dint_hdist.argmax(), Dint_hdist.shape)
Dint_points = [Dint_Dmin_myslice.vertices[Dint_bestpair[0]], Dint_Dmin_myslice.vertices[Dint_bestpair[1]]]
Dint_midpoint = (Dint_points[0] + Dint_points[1])/2
Dmax_Dmin_plane_nvec = Dint_points[0] - Dint_points[1]

#slice convex hull along Dmax/Dmin plane
Dmax_Dmin_myslice = mymesh.section(plane_origin=Dmax_midpoint, 
                        plane_normal=Dmax_Dmin_plane_nvec)
slice_2D, to_3D = Dmax_Dmin_myslice.to_planar()
#order 2D slice points by their azimuth
azimuth = np.arctan2(slice_2D.vertices[:,0], slice_2D.vertices[:,1])
sort_ind = np.argsort(azimuth)
slice_2D.vertices = slice_2D.vertices[sort_ind, :]
#calculate range (to find minimums
x_mean = np.mean(slice_2D.vertices[:,0])
y_mean = np.mean(slice_2D.vertices[:,1])
Dmax_Dmin_range = np.sqrt((slice_2D.vertices[:,0] - x_mean)**2 + (slice_2D.vertices[:,1] - y_mean)**2)
#smooth
Dmax_Dmin_range_smooth = moving_average(Dmax_Dmin_range, 15)
#find peaks
Dmax_Dmin_peaks_ind, prop = find_peaks(-Dmax_Dmin_range_smooth)
#MBB
Dmax_Dmin_bounding_box_stats = mbb.MinimumBoundingBox(slice_2D.vertices[Dmax_Dmin_peaks_ind,:])

fig = plt.figure(figsize=[8,8])
plt.scatter(slice_2D.vertices[:,0], slice_2D.vertices[:,1], c=Dmax_Dmin_range)
for ind in Dmax_Dmin_peaks_ind:
    plt.plot(slice_2D.vertices[:,0][ind], slice_2D.vertices[:,1][ind], 'r.')
mbb_points = np.array(list(Dmax_Dmin_bounding_box_stats.corner_points))
plt.plot(mbb_points[:,0], mbb_points[:,1], 'b.')

#get underlying points
internal_Dmax = Dmax_Dmin_bounding_box_stats.length_parallel
internal_Dmin1 = Dmax_Dmin_bounding_box_stats.length_orthogonal
internal_Dint = Dint_Dmin_bounding_box_stats.length_parallel
internal_Dmin2 = Dint_Dmin_bounding_box_stats.length_orthogonal
    
```
